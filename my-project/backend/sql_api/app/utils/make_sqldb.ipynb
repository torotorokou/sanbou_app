```python
import pandas as pd
import glob
import os

# ukeireディレクトリのパス
ukeire_dir = "/app/app/data/ukeire/"

# ukeire内の全csvファイルを取得
csv_files = glob.glob(os.path.join(ukeire_dir, "*.csv"))

# 全csvを読み込んでリスト化
df_list = [pd.read_csv(f) for f in csv_files]

# 全てのDataFrameを縦方向に結合（カラムが異なる場合はNaNで補完）
df = pd.concat(df_list, ignore_index=True)

# 結果のDataFrameを表示
df.head()
print(len(df))
```

```python
print(df.columns)  # カラム名一覧を表示
```

```python
import pandas as pd
import yaml
from shared.utils.dataframe_utils import remove_weekday_parentheses, remove_commas_and_convert_numeric,convert_df_by_types,combine_date_and_time

# 日付変更
target_columns = ["伝票日付","売上日付","支払日付"]

for column in target_columns:
    df = remove_weekday_parentheses(df,column)
df
```

```python
# 数値内のカンマ削除
target_columns = ["正味重量","数量","金額","単価",]

for column in target_columns:
    df = remove_commas_and_convert_numeric(df,column)
df
```

```python


# 時間型に日数を追加
target_columns = ["計量時間（空車重量）", "計量時間（総重量）"]
time_column = "伝票日付"

for column in target_columns:
    df = combine_date_and_time(df, date_col=column, time_col=time_column, new_col = column)

df
# 型変換

```

```python

# YAMLファイルの読み込み
with open('/app/config/header_types/syogun_csv.yaml', encoding='utf-8') as f:
    yaml_data = yaml.safe_load(f)

column_types = yaml_data['受入一覧']['column_types']

# 型変換の実行
df = convert_df_by_types(df, column_types)
df
```
